```python
# ------------------------------------------------
# üîç Basic Structure and Missing Overview
# ------------------------------------------------
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.3f}'.format)

# Check structure
print("‚úÖ Shape:", df.shape)
print("\nüß© Data Types:")
print(df.dtypes)

print("\nüîç Missing Values (%):")
missing = df.isnull().mean().sort_values(ascending=False) * 100
print(missing[missing > 0].round(2))
```

```python
# ------------------------------------------------
# üßπ Fill missing values and recheck
# ------------------------------------------------

# 1Ô∏è‚É£ Fill numeric NaNs with 0
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[num_cols] = df[num_cols].fillna(0)

# 2Ô∏è‚É£ Fill categorical NaNs with "Unknown"
cat_cols = df.select_dtypes(include=["object"]).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")

# 3Ô∏è‚É£ Verify again
print("‚úÖ Missing values filled successfully.\n")

print("üîç Missing values after cleaning:")
missing_post = df.isnull().mean().sort_values(ascending=False) * 100
print(missing_post[missing_post > 0].round(2))

print("\nüßÆ Final data type summary:")
print(df.dtypes.value_counts())
```

```python
# ------------------------------------------------
# ‚ö†Ô∏è Identify negative or abnormal numeric values
# ------------------------------------------------

# Columns that should never be negative
non_negative_cols = [
    "ANSR / Tech Revenue FYTD", "Billed Fees FYTD", "Billed Expenses FYTD",
    "FYTD Collections", "FYTD Labor Costs", "ETD Total Collections",
    "Total Billed FYTD", "Charged Hours FYTD", "Billed Revenue Days"
]

neg_counts = {col: (df[col] < 0).sum() for col in non_negative_cols}
neg_counts = {k: v for k, v in neg_counts.items() if v > 0}

if neg_counts:
    print("‚ö†Ô∏è Columns containing negative values:")
    for col, cnt in neg_counts.items():
        print(f"  {col}: {cnt:,} negative entries")
else:
    print("‚úÖ No negative or invalid values found in key numeric fields.")
```

```python
# ------------------------------------------------
# üè∑Ô∏è Safely flag negative values (no replacement)
# ------------------------------------------------

key_cols = [
    "ANSR / Tech Revenue FYTD", "Billed Fees FYTD", "Billed Expenses FYTD",
    "FYTD Collections", "FYTD Labor Costs", "ETD Total Collections",
    "Total Billed FYTD", "Charged Hours FYTD", "Billed Revenue Days"
]

for col in key_cols:
    df[f"{col} NegativeFlag"] = (df[col] < 0).astype(int)

print(f"‚úÖ Negative value flags added for {len(key_cols)} key financial columns.")
print(f"üìä New shape: {df.shape}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Assuming df_positive already exists from your previous step
df_model = df_positive.copy()

# ‚úÖ Keep only FY24 & FY25 for training
df_model = df_model[df_model["Year"].isin(["FY24", "FY25"])]

# Convert categorical columns to numeric
cat_cols = ["MS", "Engagement Sub Service Line", "Client", "Location", "Engagement Partner"]
encoder = LabelEncoder()

for col in cat_cols:
    df_model[col] = df_model[col].astype(str).fillna("Unknown")
    df_model[col] = encoder.fit_transform(df_model[col])

# Define target and features
target = "ANSR / Tech Revenue FYTD"
features = [
    'Total Billed FYTD', 'Billed Fees FYTD', 'FYTD Collections',
    'FYTD Labor Costs', 'Charged Hours FYTD', 'ETD Total Collections',
    'AR Outstanding', 'Billed Expenses FYTD', 'Expenses FYTD', 'EAF Reserve FYTD',
    'MS', 'Engagement Sub Service Line', 'Client', 'Location', 'Engagement Partner', 'Week'
]

X = df_model[features]
y = df_model[target]

# ‚úÖ Train/Test split (preserve time order ‚Äî no shuffling)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print(f"‚úÖ Train samples: {len(X_train):,}, Test samples: {len(X_test):,}")
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Initialize the model
xgb_model = XGBRegressor(
    n_estimators=400,
    learning_rate=0.1,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

# Train
xgb_model.fit(X_train, y_train)

# Predict
y_pred = xgb_model.predict(X_test)

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"‚úÖ RMSE: {rmse:,.2f}")
print(f"üìà R¬≤ Score: {r2:.4f}")
```

```python
import matplotlib.pyplot as plt
import xgboost as xgb

plt.figure(figsize=(10,6))
xgb.plot_importance(xgb_model, max_num_features=10)
plt.title("Top 10 Feature Importances ‚Äî Weekly ANSR FYTD Prediction")
plt.show()
```
