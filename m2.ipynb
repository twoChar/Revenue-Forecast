```python
# ------------------------------------------------
# üìä COLUMN STRUCTURE COMPARISON ‚Äî FY24 | FY25 | FY25_FY26
# ------------------------------------------------

import pandas as pd
from IPython.display import display, Markdown

# --- Create column sets ---
set_fy24 = set(df_fy24.columns)
set_fy25 = set(df_fy25.columns)
set_fy25fy26 = set(df_fy25fy26.columns)

# --- Basic column counts ---
print("üß© COLUMN COUNTS")
print(f"FY24: {len(set_fy24)} columns")
print(f"FY25: {len(set_fy25)} columns")
print(f"FY25_FY26: {len(set_fy25fy26)} columns\n")

# --- Common columns ---
common_cols = set_fy24 & set_fy25 & set_fy25fy26
print(f"‚úÖ Common columns across all 3 files: {len(common_cols)}\n")

# Display first 20 common columns as a sample
print("üß± Sample Common Columns:")
for i, col in enumerate(sorted(list(common_cols))[:20]):
    print(f"{i+1:3d}. {col}")

# --- Unique columns per dataset ---
unique_fy24 = set_fy24 - (set_fy25 | set_fy25fy26)
unique_fy25 = set_fy25 - (set_fy24 | set_fy25fy26)
unique_fy25fy26 = set_fy25fy26 - (set_fy24 | set_fy25)

print("\n‚ö†Ô∏è Unique columns in FY24:", len(unique_fy24))
print(list(unique_fy24)[:15])  # preview first 15

print("\n‚ö†Ô∏è Unique columns in FY25:", len(unique_fy25))
print(list(unique_fy25)[:15])

print("\n‚ö†Ô∏è Unique columns in FY25_FY26:", len(unique_fy25fy26))
print(list(unique_fy25fy26)[:15])

# --- Summary table ---
summary = {
    "FY24": len(set_fy24),
    "FY25": len(set_fy25),
    "FY25_FY26": len(set_fy25fy26),
    "Common to all": len(common_cols),
    "FY24 only": len(unique_fy24),
    "FY25 only": len(unique_fy25),
    "FY25_FY26 only": len(unique_fy25fy26)
}

display(Markdown("### üìã Column Comparison Summary"))
display(pd.DataFrame([summary]))
```

```python
# ------------------------------------------------
# üîÑ MERGE FY24, FY25, FY25_FY26 USING COMMON COLUMNS ONLY
# ------------------------------------------------

# Convert common_cols to a sorted list for consistent ordering
common_cols_list = sorted(list(common_cols))

# Filter each DataFrame to include only these columns
df_fy24_common = df_fy24[common_cols_list].copy()
df_fy25_common = df_fy25[common_cols_list].copy()
df_fy25fy26_common = df_fy25fy26[common_cols_list].copy()

# Concatenate all 3
df = pd.concat([df_fy24_common, df_fy25_common, df_fy25fy26_common], ignore_index=True)

# ------------------------------------------------
# ‚úÖ Check results
# ------------------------------------------------
print(f"‚úÖ Merged dataset created successfully ‚Üí shape: {df.shape}")
print("\nüìä Year distribution:")
print(df['Year'].value_counts(dropna=False))

print("\nüìã Sample data:")
display(df.head(5))
```

```python
def load_large_csv(file_path, encoding, sep, chunksize=100000):
    """Read large CSV safely using chunks."""
    chunks = []
    total_rows = 0
    for chunk in tqdm(pd.read_csv(file_path, encoding=encoding, sep=sep, chunksize=chunksize, engine="python"),
                      desc=f"Loading {os.path.basename(file_path)}"):
        chunks.append(chunk)
        total_rows += len(chunk)
    df = pd.concat(chunks, ignore_index=True)
    display(Markdown(f"‚úÖ Loaded `{os.path.basename(file_path)}` ‚Üí shape: **{df.shape}**"))
    return df

def show_data_summary(df, name="DataFrame"):
    """Display shape, sample, info, missing %, and numeric stats."""
    display(Markdown(f"### üìä {name} Overview"))
    display(df.head(3))
    display(Markdown(f"**Shape:** {df.shape[0]:,} rows √ó {df.shape[1]:,} columns"))
    display(Markdown("#### üîç Data Info"))
    display(df.info(verbose=False))
    display(Markdown("#### üßÆ Missing Values (%)"))
    display((df.isnull().mean() * 100).round(2).sort_values(ascending=False).head(10))
    display(Markdown("#### üìà Numeric Columns Summary"))
    display(df.describe(include='number').T.head(10))
```
